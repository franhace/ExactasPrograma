```python
# 12 ¿es cierto que...
# los hinchas de Boca son la mitad más uno?
import math
df_hinchas = df[df['Cuadro'].notna()] # nos quedamos con los no nan
hinchas_totales = len(df_hinchas.index)
hinchas_boca = df_hinchas[df_hinchas['Cuadro']=='Boca Juniors']
num_hinchas_boca = len(hinchas_boca.index)

num_hinchas_boca >= math.floor(hinchas_totales/2 + 1)
```

```python
# 25
carreras
pabellon_1 = ['Ciencias de la Computación','Ciencias Físicas','Ciencias Matemáticas']
pabellon_2 = []
for carrera in carreras:
    if carrera not in pabellon_1 and isinstance(carrera, str) :
        pabellon_2.append(carrera)
pabellon_2.remove('Otra')
pabellon_2
```

```python
def dame_pabellon(carrera):
    if carrera in pabellon_1:
        return 1
    elif carrera in pabellon_2:
        return 2
    else:
        return 0
dame_pabellon('Ciencias de la Atmósfera')
```

```python
df_corregido = df_pesos_corregidos.copy()
df_corregido['Carrera'] = df_corregido['Carrera'].apply(dame_pabellon)
df_corregido
```

```python
grid = sns.FacetGrid(df_corregido, col='Carrera', margin_titles=True,)
grid.map(sns.boxplot, "Promedio")
# La media de promedios es mas alta para el pabellon 2, con un de menor y mas outliers
```
